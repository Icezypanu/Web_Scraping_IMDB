```python
page_1 = 'https://www.imdb.com/search/title/?at=0&count=100&keywords=superhero&num_votes=3000,&sort=user_rating&title_type=feature'
page_2 = 'https://www.imdb.com/search/title/?title_type=feature&num_votes=3000,&keywords=superhero&sort=user_rating,desc&count=100&start=101&ref_=adv_nxt'
page_3 = 'https://www.imdb.com/search/title/?title_type=feature&num_votes=3000,&keywords=superhero&sort=user_rating,desc&count=100&start=201&ref_=adv_nxt'
page_4 = 'https://www.imdb.com/search/title/?title_type=feature&num_votes=3000,&keywords=superhero&sort=user_rating,desc&count=100&start=301&ref_=adv_nxt'
website = [page_1, page_2, page_3, page_4]
```

```python
lst_name = []
lst_year = []
lst_rating = []
lst_story = []
lst_certificate = []
lst_runtime = []
lst_genre = []
lst_director = []
lst_stars = []
lst_gross = []
lst_vote = []

for page in website:
  headers = {'Accept-Language': 'en-US,en;q=0.5'}
  req = requests.get(page, headers=headers)
  soup = BeautifulSoup(req.text, 'html.parser')

  titles = soup.find_all('div',attrs = {'class':'lister-item mode-advanced'})

  for a in titles:
    getdata = a.find('div', attrs={'class':'lister-item-content'})
    header = getdata.find('h3', attrs={'class':'lister-item-header'})
    name = header.find('a').text
    year = header.find('span', attrs={'class':"lister-item-year text-muted unbold"}).text

    text_muted = getdata.find_all('p')
    if text_muted[0].find('span', attrs={'class':"certificate"}) is None:
      certificate = 'Not Rated'
    else:
      certificate = text_muted[0].find('span', attrs={'class':"certificate"}).text

    runtime = text_muted[0].find('span', attrs={'class':"runtime"}).text
    genre = text_muted[0].find('span', attrs={'class':"genre"}).text.strip('\n').strip(' ')
    story = text_muted[1].text.strip('\n')


    rating_bar = getdata.find('div', attrs={'class':'ratings-bar'})
    rating = rating_bar.find('div', attrs={'name':'ir'}).text.strip('\n')
    director = text_muted[2].text.strip('\n').strip(' ').split('\n')[1]
    stars = text_muted[2].text.strip('\n').strip(' ').split('\n')[4:]

    vote = text_muted[3].text.strip('\n').split('\n')[1]

    if len(text_muted[3].text.strip('\n').split('\n')) < 3:
      gross = 99
    else:
      gross = text_muted[3].text.strip('\n').split('\n')[3]

    lst_name.append(name)
    lst_year.append(year)
    lst_certificate.append(certificate)
    lst_runtime.append(runtime)
    lst_genre.append(genre)
    lst_story.append(story)
    lst_rating.append(rating)
    lst_director.append(director)
    lst_stars.append(stars)
    lst_vote.append(vote)
    lst_gross.append(gross)
```

```python
print(len(lst_name), len(lst_year), len(lst_certificate), len(lst_runtime), len(lst_genre), len(lst_story), len(lst_rating), len(lst_director), len(lst_stars))
```

```python
dict_im = {'name': lst_name,
           'year': lst_year,
           'certificate': lst_certificate,
           'runtime': lst_runtime,
           'genre': lst_genre,
           'story': lst_story,
           'rating': lst_rating,
           'director': lst_director,
           'stars': lst_stars,
           'vote': lst_vote,
           'gross': lst_gross}
```

```python
import pandas as pd

df = pd.DataFrame(dict_im)
```
